```python
print(df.shape)
sns.heatmap(df.isnull(),yticklabels=False,cmap='viridis') 
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()

##the above two tables have two similar columns which are country code and country. thus we merge 
##the above two tables
```

```python
##first table has more column thus we focus on the left column that is meaning of how i.e left is main
##table and right is stuck on the left. on means which column we will stick the tables.

final_df=pd.merge(df,df_country,on='Country Code',how='left' )
```

```python
final_df.head()
```

```python
sizes=final_df.Country.value_counts().index
sizes
```

```python
## plotting a pie chart

labels=final_df.Country.value_counts().index
values=final_df.Country.value_counts().values
```

```python
##for better representation we take the largest 3 values

plt.pie(values[:3],labels=labels[:3],autopct='%1.2f%%')
```

ZOMATO IS MAJORLY OPERATING IN INDIA FOLLOWED BY USA AND UK. IT HAS MAJORITY TRANSACTIONS IN INDIA

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'rating count'})
```

# conclusions for ratings

1. when rating is between 4.5 to 4.9 then rating is excellent.
2. when rating is between 4 to 4.5 then rating is very good.
3. when rating is between 3.5 to 4.0 then rating is good.
4. when rating is between 2.5 to 3.5 then rating is average.
5. when rating is between 1.8 to 2.4 then rating is poor.



```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(14,7)
sns.barplot(x='Aggregate rating',y='rating count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='rating count',data=ratings,hue='Rating color',palette=['blue','red','orange','green','darkgreen'])
```

## observation

2. Maximum rating count is in between 2.8 to 4
